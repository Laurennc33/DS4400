#### Problem 4.2

```python
# Problem 4.2: Generate 3 random 3x3 matrices and compute inverses and product

import numpy as np

np.random.seed()  

matrices = []
inverses = []
products = []

for i in range(3):
    while True:
        M = np.random.randint(-10, 11, size=(3,3))
        det = np.linalg.det(M)
        if abs(det) > 1e-8:  
            break
    matrices.append(M)
    inv_M = np.linalg.inv(M)
    inverses.append(inv_M)
    products.append(np.dot(M, inv_M))

# Display results
for i in range(3):
    print(f"\nMatrix {i+1}:")
    print(matrices[i])
    
    print("\nInverse:")
    print(inverses[i])
    
    print("\nProduct (Matrix * Inverse):")
    print(products[i])
    print("\n" + "-"*50)
```

#### Problem 5.1

```python
import pandas as pd

# Load dataset 
df = pd.read_csv("kc_house_data.csv")

# Drop columns to ignore
df = df.drop(columns=['id', 'date', 'zipcode'])

# Compute statistics
stats = pd.DataFrame({
    'mean': df.mean(),
    'min': df.min(),
    'max': df.max(),
    'variance': df.var()
})

print("Feature statistics (average, min, max, variance):")
print(stats)

# Find feature with lowest and highest average
lowest_avg_feature = stats['mean'].idxmin()
highest_avg_feature = stats['mean'].idxmax()

print("\nLowest average:")
print(f"{lowest_avg_feature}: {stats.loc[lowest_avg_feature, 'mean']}")

print("\nHighest average:")
print(f"{highest_avg_feature}: {stats.loc[highest_avg_feature, 'mean']}")

# Find feature with lowest and highest variance
lowest_var_feature = stats['variance'].idxmin()
highest_var_feature = stats['variance'].idxmax()

print("\nLowest variance:")
print(f"{lowest_var_feature}: {stats.loc[lowest_var_feature, 'variance']}")

print("\nHighest variance:")
print(f"{highest_var_feature}: {stats.loc[highest_var_feature, 'variance']}")
```

#### Problem 5.2

```python
# Compute correlation of each feature with the response 

import pandas as pd

# Load dataset
df = pd.read_csv('kc_house_data.csv')

# Drop columns to ignore
df = df.drop(columns=['id', 'date', 'zipcode'])

# Compute correlations with price
correlations = df.corr()['price'].drop('price')

# Convert to DataFrame for clean table
corr_table = correlations.sort_values(ascending=False).reset_index()
corr_table.columns = ['Feature', 'Correlation_with_Price']

corr_table
```

All of the features have positive correlation coefficients, which means as the feature increases, on average house price tends to increase. The feature which has the highest positive correlation is sqft_living at 0.7020. This means this feature is the most important predictor in terms of linear relationship with price.
